## អថេរចៃដន្យ និងការបែងចែក
ចាប់ផ្តើមដោយគូសគំរូតម្លៃ ៣០ ពីការបែងចែកសមាមាត្រពី ០ ដល់ ៩។ យើងនឹងគណនា​មធ្យម និងអាំងធរីយ៍ផងដែរ។

## ចន្លោះទំនុកចិត្ត

ឥឡូវនេះយើងនឹងគណនាចន្លោះទំនុកចិត្តសម្រាប់ទម្ងន់និងកម្ពស់របស់អ្នកលេងបេស្បាល់។ យើងនឹងប្រើកូដ [ពីការពិភាក្សានៅ stackoverflow នេះ](https://stackoverflow.com/questions/15033511/compute-a-confidence-interval-from-sample-data):


```python
import scipy.stats

def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, h

for p in [0.85, 0.9, 0.95]:
    m, h = mean_confidence_interval(df['Weight'].fillna(method='pad'),p)
    print(f"p={p:.2f}, mean = {m:.2f} ± {h:.2f}")
```

## ការធ្វើតេស្តសន្និដ្ឋាន

ចូរយើងស្រាវជ្រាវតួនាទីផ្សេងៗនៅក្នុងឃ្លាំងទិន្នន័យកីឡាករបេស្បលរបស់យើង៖


```python
df.groupby('Role').agg({ 'Weight' : 'mean', 'Height' : 'mean', 'Age' : 'count'}).rename(columns={ 'Age' : 'Count'})
```

យើង​ចង់​សាកល្បង​ការ​សន្មត់​ថាអ្នក​ធ្វើ​ងារ​ដំបូងមានកម្ពស់​ខ្ពស់ជាង​អ្នក​ធ្វើ​ងារ​ទីពីរ។ វិធីសាមញ្ញបំផុតក្នុងការធ្វើនេះគឺការតេស្តចន្លោះទំនុកចិត្តៈ


```python
for p in [0.85,0.9,0.95]:
    m1, h1 = mean_confidence_interval(df.loc[df['Role']=='First_Baseman',['Height']],p)
    m2, h2 = mean_confidence_interval(df.loc[df['Role']=='Second_Baseman',['Height']],p)
    print(f'Conf={p:.2f}, 1st basemen height: {m1-h1[0]:.2f}..{m1+h1[0]:.2f}, 2nd basemen height: {m2-h2[0]:.2f}..{m2+h2[0]:.2f}')
```

យើងអាចឃើញថា ខ្សែពេលពុំមានការឆ្លងគ្នាទេ។

វិធីសាស្រ្តដែលត្រឹមត្រូវជាងសម្រាប់បញ្ជាក់សន្មត់គឺការប្រើប្រាស់ **មេត្រការសាកល្បង t របស់សិស្ស**:


```python
from scipy.stats import ttest_ind

tval, pval = ttest_ind(df.loc[df['Role']=='First_Baseman',['Height']], df.loc[df['Role']=='Second_Baseman',['Height']],equal_var=False)
print(f"T-value = {tval[0]:.2f}\nP-value: {pval[0]}")
```

ពីរតម្លៃដែលត្រូវបានត្រឡប់តាមរយៈមុខងារ `ttest_ind` គឺ៖
* p-value អាចត្រូវបានគិតជាប្រាស្រ័យភាពនៃការចែកចាយពីរដែលមានមធ្យមដូចគ្នា។ សម្រាប់ករណីរបស់យើង វាទាបណាស់ មានន័យថាមានភស្តុតាងខ្លាំងគាំទ្រថា first basemen មានកម្ពស់ខ្ពស់ជាង។
* t-value គឺជាតម្លៃមន្ទីរប្រមាណនៃភាពខុសគ្នាមធ្យមដែលបានធម្មតា ដែលត្រូវបានប្រើនៅក្នុងការស្ទង់មតិ t-test ហើយវាត្រូវបានប្រៀបធៀបទៅនឹងតម្លៃកម្រិតមួយសម្រាប់តម្លៃទំនុកចិត្តមួយ។


## ការសម្រួលចំណែកធម្មជាតិស្របតាមវឌ្ឍនភាពកណ្តាល

ម៉ាស៊ីនបង្កើតចៃដន្យក្នុង Python ត្រូវបានរចនាឡើងដើម្បីផ្តល់នូវចំណែកសាមញ្ញតែមួយ។ ប្រសិនបើយើងចង់បង្កើតម៉ាស៊ីនបង្កើតសម្រាប់ចំណែកធម្មជាតិ យើងអាចប្រើវឌ្ឍនភាពកណ្តាល។ ដើម្បីទទួលបានតម្លៃចែកចាយធម្មជាតិ យើងអាចគណនាមធ្យមនៃចំណែកដែលបង្កើតដោយសាមញ្ញ។


```python
def normal_random(sample_size=100):
    sample = [random.uniform(0,1) for _ in range(sample_size) ]
    return sum(sample)/sample_size

sample = [normal_random() for _ in range(100)]
plt.figure(figsize=(10,6))
plt.hist(sample)
plt.tight_layout()
plt.show()
```

ការតភ្ជាប់អនុញ្ញាតឲ្យយើងស្វែងរកទំនាក់ទំនងរវាងលំដាប់ទិន្នន័យ។ នៅក្នុងឧទាហរណ៍ក្មេងលេងរបស់យើង, សូមចរិតថាមានក្រុមហ៊ុនបេសបូលអាក្រក់មួយដែលបង់ប្រាក់ទៅឱ្យកីឡាកររបស់ខ្លួនតាមតម្លៃកម្ពស់ - កីឡាករដែលខ្ពស់ជាងនឹងទទួលបានប្រាក់ច្រើនជាង។ សូមសន្មតថាកំណត់ប្រាក់ខែមូលដ្ឋាន $1000 ហើយមានប្រាក់រង្វាន់បន្ថែមចាប់ពី $0 ដល់ $100 អាស្រ័យលើកម្ពស់។ យើងនឹងយកកីឡាករពិតពី MLB ហើយគណនាប្រាក់ខែស្រមៃរបស់ពួកគេ៖


```python
heights = df['Height'].fillna(method='pad')
salaries = 1000+(heights-heights.min())/(heights.max()-heights.mean())*100
print(list(zip(heights, salaries))[:10])
```

ឥឡូវយើងនឹងគណនាការសហភាព និងការតភ្ជាប់គ្នារបស់ចំណាត់ថ្នាក់ទាំងនោះ។ `np.cov` នឹងផ្តល់ឲ្យយើងនូវអ្វីដែលគេហៅថា **ម៉ាត្រិកសហភាព** ដែលជាការពង្រីកនៃការសហភាពទៅច្រើនអថេរ។ គ្រឿង $M_{ij}$ នៃម៉ាត្រិកសហភាព $M$ គឺជាការតភ្ជាប់គ្នារវាងអថេរបញ្ចូល $X_i$ និង $X_j$ ហើយតម្លៃបន្ទាត់ឈរ $M_{ii}$ គឺជាការប្រែប្រួលនៃ $X_{i}$។ ដូចគ្នានេះ `np.corrcoef` នឹងផ្តល់ឲ្យយើងនូវ **ម៉ាត្រិកតភ្ជាប់គ្នា**។


```python
print(f"Covariance matrix:\n{np.cov(heights, salaries)}")
print(f"Covariance = {np.cov(heights, salaries)[0,1]}")
print(f"Correlation = {np.corrcoef(heights, salaries)[0,1]}")
```

ការតភ្ជាប់ដែលស្មើនឹង 1 មានន័យថាមាន **ទំនាក់ទំនងកែវធីត្រង់** ដែលខ្លាំងមួយរវាងអថេរពីរនេះ។ យើងអាចមើលឃើញទំនាក់ទំនងកែវធីត្រង់ដោយការគូរតម្លៃមួយប្រឆាំងនឹងមួយផ្សេងទៀតបានជាផ្ទាល់៖


```python
plt.figure(figsize=(10,6))
plt.scatter(heights,salaries)
plt.tight_layout()
plt.show()
```

មកមើលថាអ្វីនឹងកើតឡើងបើទំនាក់ទំនងមិនមែនជារស់រាយ។ សន្មតថាក្រុមហ៊ុនរបស់យើងបានសម្រេចចិត្តលាក់បាំងការពឹងផ្អែកលើរូបមន្តរាយច្បាស់រវាងកម្ពស់និងប្រាក់ខែ ហើយបានបញ្ចូលអង្គភាពមិនរាយមួយចូលក្នុងរូបមន្ត ដូចជា `sin`:


```python
salaries = 1000+np.sin((heights-heights.min())/(heights.max()-heights.mean()))*100
print(f"Correlation = {np.corrcoef(heights, salaries)[0,1]}")
```

ក្នុងករណីនេះ ការចងកCorrារតិ្តគឺតូចជាងបន្តិច ប៉ុន្តែវាទោះជាយ៉ាងណាក៏ដោយនៅតែខ្ពស់។ ឥឡូវនេះ ដើម្បីធ្វើឱ្យទំនាក់ទំនងមិនច្បាស់លាស់ យើងអាចចង់បន្ថែមភាពចៃដន្យបន្ថែមមួយចំនួនដោយបន្ថែមអថេរចៃដន្យមួយចំនួនទៅលើប្រាក់ខែ។ សូមមកមើលថាអ្វីកើតឡើង៖


```python
salaries = 1000+np.sin((heights-heights.min())/(heights.max()-heights.mean()))*100+np.random.random(size=len(heights))*20-10
print(f"Correlation = {np.corrcoef(heights, salaries)[0,1]}")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(heights, salaries)
plt.tight_layout()
plt.show()
```

> តើអ្នកអាចទាយបានទេថាមូលហេតុដែលចំណុចទាំងនេះរៀបខ្នងតាមបន្ទាត់ដេកដូចនេះ?

យើងបានសង្កេតឃើញការតភ្ជាប់រវាងគំនិតដែលបានបង្កើតឡើងដោយក្លែងបំពានដូចជា वेतन និងអថេរដែលបានសង្កេតឃើញ *កំពស់*។ តោះមើលថាតើអថេរដែលបានសង្កេតពីរដូចជា កំពស់ និង ទំងន់ មានការតភ្ជាប់គ្នាទៅដែរ៖


```python
np.corrcoef(df['Height'].ffill(),df['Weight'])
```

យ៉ាងក្រៀមក្រំពេក យើងមិនទទួលបានលទ្ធផលណាមួយទេ - មានតែតម្លៃ `nan` ដែលចម្លែកខ្លះប៉ុណ្ណោះ។ នេះបណ្តាលមកពីការពិតថាតម្លៃខ្លះនៅក្នុងស៊េរីរបស់យើងមិនបានកំណត់ សម្គាល់ជាទម្រង់ `nan` ដែលបណ្តាលឲ្យលទ្ធផលនៃប្រតិបត្តិការមិនបានកំណត់ដែរ។ ដោយមើលទៅលើម៉ាទ្រីស យើងអាចឃើញថា `Weight` គឺជាឈ្មោះជួរឈរដែលមានបញ្ហា ពីព្រោះការជាប់គ្នាទៅវិញទៅមករវាងតម្លៃ `Height` ត្រូវបានគណនា។

> ឧទាហរណ៍នេះបង្ហាញពីសារៈសំខាន់នៃ **ការរៀបចំទិន្នន័យ** និង **ការសំអាត**។ ពុំមានទិន្នន័យត្រឹមត្រូវ យើងមិនអាចគណនាអ្វីបានទេ។

ចង់យកវិធីសាស្ត្រ `fillna` ដើម្បីបំពេញតម្លៃខ្វះ ហើយគណនាការជាប់គ្នាទៀត៖ 


```python
np.corrcoef(df['Height'].fillna(method='pad'), df['Weight'])
```

ពិតជាមានសមាបាត្រ ប៉ុន្តែមិនខ្លាំងដូចក្នុងឧទាហរណ៍ក្លែងក្លាយរបស់យើងទេ។ ពិតណាស់ ប្រសិនបើយើងមើលផ្លូវកំណត់ចំណុចរវាងតម្លៃមួយនិងមួយផ្សេងទៀត ទំនាក់ទំនងនោះនឹងមិនច្បាស់លាស់ដូចមុនទេ៖

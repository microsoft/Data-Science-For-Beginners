## ယုံကြည်မှု အကြားကွာဟမှုများ

အခုတော့ ဘေ့စ်ဘောကစားသမားများ၏ အလေးချိန်နှင့် အမြင့်အတွက် ယုံကြည်မှု အကြားကွာဟမှုများကိုတွက်ချက်ကြမည်။ ကျွန်ုပ်တို့သည် [ဒီ stackoverflow ဆွေးနွေးချက်မှ](https://stackoverflow.com/questions/15033511/compute-a-confidence-interval-from-sample-data) ကိုဒ်ကိုအသုံးပြုမည်။


```python
import scipy.stats

def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, h

for p in [0.85, 0.9, 0.95]:
    m, h = mean_confidence_interval(df['Weight'].fillna(method='pad'),p)
    print(f"p={p:.2f}, mean = {m:.2f} ± {h:.2f}")
```

## Hypothesis Testing

ကျွန်ုပ်တို့၏ဘေ့စ်ဘောကစားသူများဒေတာသိုက်တွင် သုံးဆောင်သော တာဝန်များကို ရှာဖွေကြမည်။


```python
df.groupby('Role').agg({ 'Weight' : 'mean', 'Height' : 'mean', 'Age' : 'count'}).rename(columns={ 'Age' : 'Count'})
```

ပထမဦးဆုံး ဘေ့စ်မန်များသည် ဒုတိယဦးဆုံး ဘေ့စ်မန်များထက် ပိုရှည်လျားကြောင်း အယူအဆကို စမ်းသပ်ကြည့်ပါစို့။ ဤအယူအဆကို စစ်ဆေးရန် အလွယ်တကူနည်းလမ်းမှာ ယုံကြည်စိတ်ချမှုအကြောင်းအရာများကို စစ်ဆေးခြင်းဖြစ်သည်။


```python
for p in [0.85,0.9,0.95]:
    m1, h1 = mean_confidence_interval(df.loc[df['Role']=='First_Baseman',['Height']],p)
    m2, h2 = mean_confidence_interval(df.loc[df['Role']=='Second_Baseman',['Height']],p)
    print(f'Conf={p:.2f}, 1st basemen height: {m1-h1[0]:.2f}..{m1+h1[0]:.2f}, 2nd basemen height: {m2-h2[0]:.2f}..{m2+h2[0]:.2f}')
```

အစီအစဉ်များသည် ထပ်တွဲမှုမရှိကြောင်း တွေ့မြင်ရပါသည်။

သရုပ်ပြချက်ကို ပိုမိုမှန်ကန်စွာ သက်သေပြရန် အဆင့်မြင့် ထုတ်ပြန်မှုမှာ **Student t-test** ကို အသုံးပြုခြင်းဖြစ်သည်။


```python
from scipy.stats import ttest_ind

tval, pval = ttest_ind(df.loc[df['Role']=='First_Baseman',['Height']], df.loc[df['Role']=='Second_Baseman',['Height']],equal_var=False)
print(f"T-value = {tval[0]:.2f}\nP-value: {pval[0]}")
```

`ttest_ind` function က ပြန်လာတဲ့ တန်ဖိုး နှစ်ခုကတော့ -
* p-value က နှစ်ခုရဲ့ distribution တူညီတဲ့ mean တန်ဖိုးကြားရှိနိုင်မှု ရာခိုင်နှုန်းအနေနဲ့ယူဆနိုင်ပါတယ်။ ကျွန်တော်တို့မှာတော့ ကျယ်လွန်တဲ့ အထောက်အထားတွေ ရှိတယ်ဆိုတာကို ရှင်းပြတာပါ၊ ဒါကြောင့် ပထမဆုံး baseman တွေဟာ ပိုပြီးပါးစပ်ကြီးတယ်လို့ ဆိုနိုင်ပါတယ်။
* t-value က t-test တွင် အသုံးပြုတဲ့ normalized mean difference ရဲ့ အလယ်အလတ်တန်ဖိုးဖြစ်ပြီး အဆိုပါ confidence value အတွက် သတ်မှတ်ထားတဲ့ နံပါတ်တန်ဖိုးနဲ့ နှိုင်းယှဉ်ကြည့်ပါတယ်။


## Central Limit Theorem ဖြင့် ပုံမှန်ဖြန့်ဝေမှုကို နမူနာဖန်တီးခြင်း

Python မှာ pseudo-random generator ကို uniform distribution ပေးဖို့ ဒီဇိုင်းလုပ်ထားပါတယ်။ ပုံမှန်ဖြန့်ဝေမှုအတွက် generator ဖန်တီးချင်ရင်၊ central limit theorem ကို အသုံးပြုနိုင်ပါတယ်။ ပုံမှန်ဖြန့်ဝေမှုတန်ဖိုးကို ရဖို့ uniform-generated နမူနာတစ်ခုရဲ့ အလယ်တန်းကိုတွက်ချက်ပေးရမယ်။


```python
def normal_random(sample_size=100):
    sample = [random.uniform(0,1) for _ in range(sample_size) ]
    return sum(sample)/sample_size

sample = [normal_random() for _ in range(100)]
plt.figure(figsize=(10,6))
plt.hist(sample)
plt.tight_layout()
plt.show()
```

## ဆက်စပ်မှုနှင့် မကောင်းသော ဘေ့စ်ဘော ကော်ပိုရေးရှင်း

ဆက်စပ်မှုသည် ဒေတာအစဉ်များအကြား ဆက်နွယ်မှုများကို ရှာဖွေရန် အခွင့်အလမ်းပေးသည်။ ကျွန်ုပ်တို့၏ ကစားစက်ဥပမာတွင် မကောင်းသော ဘေ့စ်ဘော ကော်ပိုရေးရှင်းတစ်ခုရှိပြီး ၎င်းသည် ကစားသမားများဆီကို အမြင့်အလိုက် လစာပေးနေသည်ဟု သဘောထားကြရအောင် - ကစားသမားမှာ မြင့်မားသမျှ အပိုငွေများကို ရရှိမည်။ မူလလစာ $1000 ရှိပြီး၊ အမြင့်ပမာဏအပေါ်မူတည်၍ $0 မှ $100 အထိ အပိုအားသာငွေ ရရှိနိုင်သည် ဟု သတ်မှတ်ကြသည်။ ကျွန်ုပ်တို့ MLB မှ တကယ့်ကစားသမားများကို ယူပြီး သူတို့၏ ဟောင်းသော လစာများကိုတွက်မည်။


```python
heights = df['Height'].fillna(method='pad')
salaries = 1000+(heights-heights.min())/(heights.max()-heights.mean())*100
print(list(zip(heights, salaries))[:10])
```

ယခုဆိုရင် ဤအဆက်စပ်များ၏ ကိုဗားရီးယားနှင့် ကိုယ်စားပြုချက်အား တွက်ချက်ကြမည်။ `np.cov` သည် ကျွန်ုပ်တို့အား အဆိုပါ **ကိုဗားရီးယား မက်ထရစ်စ်** ကို ပေးမည်ဖြစ်ပြီး၊ ၎င်းမှာ များပြားသော မည့်သည့် ပြောင်းလဲမှုများအတွက် ကိုဗားရီးယား၏ ဖြည်းဖြည်းချင်း စွဲသုံးမှုတစ်ခုဖြစ်ပါသည်။ ကိုဗားရီးယား မက်ထရစ်စ် $M$ ၏ အတိုင်းအတာ $M_{ij}$ သည် ထည့်သွင်းမည့် ပြောင်းလဲမှုများ $X_i$ နှင့် $X_j$ တို့အကြား ကိုယ်စားပြုချက်ဖြစ်ပြီး၊ ကွက်တိပေါ်ရှိ တန်ဖိုးများ $M_{ii}$ သည် $X_{i}$ ၏ မတည်ငြိမ်မှုဖြစ်သည်။ သဘောတူညီ၍ `np.corrcoef` သည် ကျွန်ုပ်တို့အား **ကိုယ်စားပြုချက် မက်ထရစ်စ်** ကို ပေးမည်ဖြစ်သည်။


```python
print(f"Covariance matrix:\n{np.cov(heights, salaries)}")
print(f"Covariance = {np.cov(heights, salaries)[0,1]}")
print(f"Correlation = {np.corrcoef(heights, salaries)[0,1]}")
```

ကိုယ်စားပြုတန်ဖိုး 1 နှင့်တူညီသည်ဆိုသည်မှာ အမျိုးအစားနှစ်ခုအကြား ပြင်းထန်သော **လိုင်နာဆက်စပ်မှု** ရှိသည်ကို ဆိုလိုသည်။ တန်ဖိုးတစ်ခုအား တစ်ခုကို အဆင့်လိုက်ထိုးပြခြင်းဖြင့် လိုင်းဆက်စပ်မှုကို ကိုယ်ရည်အသိပိုင်းဖြင့် တွေ့မြင်နိုင်ပါသည်။


```python
plt.figure(figsize=(10,6))
plt.scatter(heights,salaries)
plt.tight_layout()
plt.show()
```

အဆက်အဆံမျိုးသည် လိုင်းနစ်မဟုတ်ပါက ဘာတွေဖြစ်မလဲဆိုတာ ကြည့်ကြရအောင်။ ကျွန်ုပ်တို့၏ကုမ္ပဏီသည် အမြင့်နှင့် လစာအကြား ရိုးရှင်းသော လိုင်းနစ်ဆက်နှင့်မှုကို ဖုံးကွယ်ရန်ဆုံးဖြတ်ခဲ့ပြီး၊ `sin` ကဲ့သို့သော အမျိုးအစား မလိုင်းနစ်မျိုးကို ဖော်မြူလာထဲသို့ ထည့်သွင်းခဲ့သည်ဟု ယူဆပါစို့။


```python
salaries = 1000+np.sin((heights-heights.min())/(heights.max()-heights.mean()))*100
print(f"Correlation = {np.corrcoef(heights, salaries)[0,1]}")
```

ဒီအခါမှာ ဆက်စပ်မှုက စာနည်းနည်းသာ လျော့နည်းနေတာ ဖြစ်ပေမယ့် အရမ်းမြင့်မားနေဆဲ ဖြစ်ပါတယ်။ အခုတော့ ဆက်စပ်မှုကို နည်းနည်း မသေချာအောင်လုပ်ဖို့ အလုပ်ခကို စိတ်ကူးမဲ့ အမျိုးအစားတစ်ခု ထည့်ပြီး အလှည့်အပြောင်း ပိုမိုတွေ့လာစေချင်တယ်။ ဘာဖြစ်လာမလဲ ကြည့်ကြရအောင်။


```python
salaries = 1000+np.sin((heights-heights.min())/(heights.max()-heights.mean()))*100+np.random.random(size=len(heights))*20-10
print(f"Correlation = {np.corrcoef(heights, salaries)[0,1]}")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(heights, salaries)
plt.tight_layout()
plt.show()
```

> အကြောင်းမှာ အမှတ်လေးတွေက ဒီလို ထောင့်စောင်းလိုက်တယ်ဆိုတာ ကို နှုတ်ဆက်မtightရနိုင်လား?

ကျွန်ုပ်တို့သည် လစာကဲ့သို့ အသုံးပြုမှုဖြင့် ထိန်းချုပ်ထားသောအကြောင်းအရာနှင့် ကြည့်ရှုသောအခြေအနေ* အမြင့်* ၏တွဲဖက်မှုကို တွေ့ရှိခဲ့ပြီးဖြစ်သည်။ ထို့အပြင်၊ အမြင့်နှင့် ဝန်အားကဲ့သို့ ကြည့်ရှုသောအခြေအနေ နှစ်ခုလုံးအကြားလည်း တွဲဖက်မှုရှိမရှိကိုကြည့်လိုက်ကြပါစို့။


```python
np.corrcoef(df['Height'].ffill(),df['Weight'])
```

ကံမကောင်းစွာဖြင့် ကျွန်ုပ်တို့ အရာများအတွက် ရလဒ် မရရှိကြပါဘူး - အတိုင်းအတာ `nan` အနည်းငယ်သာ ရှိနေပါတယ်။ ၎င်းမှာဖြစ်ရခြင်းမှာ ကျွန်ုပ်တို့၏ series အချို့၏ တန်ဖိုးများသည် အမည်မသိဖြစ်၍ `nan` အဖြစ် ဖော်ပြထားသလို ဖြစ်သည်၊ ၎င်းကြောင့် အလုပ်လုပ်ပုံရလဒ်လည်း အမည်မသိ ဖြစ်နေသည်။ matrix ကို ကြည့်မယ်ဆိုရင် `Weight` ကန့်သတ်သော ကော်လံဖြစ်ကာ၊ `Height` တန်ဖိုးများအကြား ကိုယ်တိုင် အချိုးအစားချက်ကို တွက်ချက်ပြီးဖြစ်သည်ကို မြင်ရသည်။

> ဤနမူနာသည် **ဒေတာပြင်ဆင်ခြင်း** နှင့် **ရှင်းလင်းခြင်း** ၏ အရေးကြီးမှုကို ပြသသည်။ သင့်တော်သော ဒေတာ မရှိဘဲနဲ့ ကိစ္စရာ များကို တွက်ချက်၍ မရနိုင်ပါ။

`fillna` နည်းလမ်းကို အသုံးပြု၍ လိုအပ်သော တန်ဖိုးများကို ဖြည့်ပြီး အချိုးအစားချက်ကို တွက်ချက်ကြမယ်။


```python
np.corrcoef(df['Height'].fillna(method='pad'), df['Weight'])
```

တကယ်တော့ ဆက်စပ်မှုတစ်ခုရှိပါသည်၊ သို့သော် ကျွန်ုပ်တို့လုပ်ဆောင်ထားသော လူမှုရိုက်ထွဲထားသော ဥပမာကဲ့သို့ ပြင်းထန်မှု မရှိပါ။ တကယ်လို့ တန်ဖိုးတစ်ဖက်ကို တစ်ဖက်နှင့် ဆက်စပ်မှုရှိမှုကို scatter plot မှ ဖြတ်သွားကြည့်ပါက၊ ဆက်စပ်မှုမှာ ပိုသေးငယ်ပြီး မမြင်သာနိုင်ပါ။


```python
plt.figure(figsize=(10,6))
plt.scatter(df['Weight'],df['Height'])
plt.xlabel('Weight')
plt.ylabel('Height')
plt.tight_layout()
plt.show()
```

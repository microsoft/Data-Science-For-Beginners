# تجسم تولید عسل 🍯 🐝


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
honey = pd.read_csv('../../../data/honey.csv')
honey.head()
```

برای نشان دادن رابطه بین یک ایالت و قیمت هر پوند برای عسل محلی از نمودار پراکندگی استفاده کنید


```python
sns.relplot(x="priceperlb", y="state", data=honey, height=15, aspect=.5);
```

این رابطه را به همراه یک نمودار پراکندگی جدید با رنگ زرد نشان دهید که رنگ آن بر اساس سال تعیین شده است.


```python
sns.relplot(x="priceperlb", y="state", hue="year", palette="YlOrBr", data=honey, height=15, aspect=.5);
```

این نمودار را یک بار دیگر امتحان کنید، این بار تغییر قیمت را با اندازه نقطه در نمودار پراکندگی نشان دهید


```python
sns.relplot(x="priceperlb", y="state", size="year", data=honey, height=15, aspect=.5);
```

برخی گزینه‌های نمودار خطی را بررسی کنید


```python
sns.relplot(x="year", y="priceperlb", kind="line", data=honey);
```

```python
sns.relplot(x="year", y="totalprod", kind="line", data=honey);
```

ایجاد یک شبکه وجه


```python

sns.relplot(
    data=honey, 
    x="yieldpercol", y="numcol",
    col="year", 
    col_wrap=3,
    kind="line"
)
```

یک نمودار خطی دوگانه ایجاد کنید (این راه‌حل توسط کدار غوله پیشنهاد شده است: https://kedar.hashnode.dev/how-to-combine-two-line-charts-in-seaborn-and-python)


```python
fig, ax = plt.subplots(figsize=(12,6))
lineplot = sns.lineplot(x=honey['year'], y=honey['numcol'], data=honey, 
                        label = 'Number of bee colonies', legend=False)
sns.despine()
plt.ylabel('# colonies')
plt.title('Honey Production Year over Year');

ax2 = ax.twinx()
lineplot2 = sns.lineplot(x=honey['year'], y=honey['yieldpercol'], ax=ax2, color="r", 
                         label ='Yield per colony', legend=False) 
sns.despine(right=False)
plt.ylabel('colony yield')
ax.figure.legend();
```

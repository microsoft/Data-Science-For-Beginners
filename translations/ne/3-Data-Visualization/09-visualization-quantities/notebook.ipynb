# चरा सम्बन्धी कुरा सिकौं

## चरा के हो?

चरा पखेटा भएका, प्वाँखले ढाकिएका, र प्रायः उड्न सक्ने जीव हुन्। तिनीहरू अण्डा पार्छन् र संसारभरि पाइन्छन्। 

## चराहरूका विशेषताहरू

- **प्वाँख**: प्वाँखले चराहरूलाई न्यानो राख्छ र उड्न मद्दत गर्छ।  
- **हल्का हड्डी**: तिनीहरूको हड्डी हल्का हुन्छ, जसले उड्न सजिलो बनाउँछ।  
- **चोच**: चरा खाना खान र विभिन्न काम गर्न चोच प्रयोग गर्छन्।  

## चराहरूको वर्गीकरण

चराहरूलाई विभिन्न समूहमा वर्गीकृत गर्न सकिन्छ। केही उदाहरणहरू:  
1. **गायक चरा**: जस्तै बुलबुल, काग।  
2. **सिकार गर्ने चरा**: जस्तै बाज, गिद्ध।  
3. **पानीमा बस्ने चरा**: जस्तै हाँस, बकुल्ला।  

## चरा किन महत्त्वपूर्ण छन्?

चराहरूले पारिस्थितिक प्रणालीमा महत्त्वपूर्ण भूमिका खेल्छन्।  
- तिनीहरूले कीरा नियन्त्रण गर्छन्।  
- फूलहरूको परागसेचनमा मद्दत गर्छन्।  
- बीउ फैलाउन सहयोग गर्छन्।  

[!TIP] चराहरूलाई नजिकबाट हेर्न चाहनुहुन्छ भने, आफ्नो बगैंचामा चराहरूलाई आकर्षित गर्न दाना राख्नुहोस्।  

## चरा संरक्षण

चराहरूको संख्या घट्दै गएको छ। तिनीहरूलाई जोगाउन हामीले के गर्न सक्छौं?  
- रूखहरू रोपौं।  
- तिनीहरूको बासस्थान नष्ट नगर्नुहोस्।  
- प्लास्टिकको प्रयोग कम गरौं।  

[!IMPORTANT] चराहरूको संरक्षण गर्नु हाम्रो जिम्मेवारी हो।  

## निष्कर्ष

चराहरू सुन्दर र उपयोगी जीव हुन्। तिनीहरूको बारेमा जान्न र तिनीहरूलाई जोगाउन हामी सबैले प्रयास गर्नुपर्छ।

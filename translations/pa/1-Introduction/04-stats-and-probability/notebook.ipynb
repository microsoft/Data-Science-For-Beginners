## ਭਰੋਸੇਯੋਗ ਇੰਟਰਵਲ

ਹੁਣ ਅਸੀਂ ਬੇਸਬਾਲ ਖਿਡਾਰੀਆਂ ਦੇ ਵਜ਼ਨ ਅਤੇ ਕਦ ਦੀਆਂ ਭਰੋਸੇਯੋਗ ਇੰਟਰਵਲ ਦੀ ਗਣਨਾ ਕਰਾਂਗੇ। ਅਸੀਂ ਇਹ ਕੋਡ [ਇਸ stackoverflow ਚਰਚਾ ਤੋਂ](https://stackoverflow.com/questions/15033511/compute-a-confidence-interval-from-sample-data) ਵਰਤਾਂਗੇ:


```python
import scipy.stats

def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, h

for p in [0.85, 0.9, 0.95]:
    m, h = mean_confidence_interval(df['Weight'].fillna(method='pad'),p)
    print(f"p={p:.2f}, mean = {m:.2f} ± {h:.2f}")
```

## ਪਰਿਕਲਪਨਾ ਦੀ ਜਾਂਚ

ਆਓ ਸਾਡੇ ਬੇਸਬਾਲ ਖਿਡਾਰੀਆਂ ਦੇ ਡੇਟਾਸੇਟ ਵਿੱਚ ਵੱਖ-ਵੱਖ ਭੂਮਿਕਾਵਾਂ ਦੀ ਖੋਜ ਕਰੀਏ:


```python
df.groupby('Role').agg({ 'Weight' : 'mean', 'Height' : 'mean', 'Age' : 'count'}).rename(columns={ 'Age' : 'Count'})
```

ਆਓ ਇਹ ਪਰਿਕਲਪਨਾ ਜਾਂਚੀਏ ਕਿ ਪਹਿਲੇ ਬੇਸਮੈਨ ਦੂਜੇ ਬੇਸਮੈਨ ਨਾਲੋਂ ਲੰਬੇ ਹੁੰਦੇ ਹਨ। ਇਸ ਦੀ ਸਭ ਤੋਂ ਸੌਖੀ ਤਰੀਕਾ ਇਹ ਹੈ ਕਿ ਕਾਂਫੀਡੈਂਸ ਇੰਟਰਨਲ ਦੀ ਜਾਂਚ ਕੀਤੀ ਜਾਵੇ:


```python
for p in [0.85,0.9,0.95]:
    m1, h1 = mean_confidence_interval(df.loc[df['Role']=='First_Baseman',['Height']],p)
    m2, h2 = mean_confidence_interval(df.loc[df['Role']=='Second_Baseman',['Height']],p)
    print(f'Conf={p:.2f}, 1st basemen height: {m1-h1[0]:.2f}..{m1+h1[0]:.2f}, 2nd basemen height: {m2-h2[0]:.2f}..{m2+h2[0]:.2f}')
```

ਅਸੀਂ ਦੇਖ ਸਕਦੇ ਹਾਂ ਕਿ ਇੰਟਰਵਲਾਂ ਇਕ ਦੂਜੇ ਨਾਲ ਓਵਰਲੈਪ ਨਹੀਂ ਕਰਦੇ। 

ਐਕ ਅੰਕੜਾ ਵਿਗਿਆਨਕ ਤੌਰ 'ਤੇ ਹੋਰ ਸਹੀ ਤਰੀਕਾ ਹੈ ਕਿ ਪਰਿਕਲਪਨਾ ਨੂੰ ਸਾਬਤ ਕਰਨ ਲਈ **ਸ੍ਟੂਡੈਂਟ t-ਟੈਸਟ** ਦੀ ਵਰਤੋਂ ਕੀਤੀ ਜਾਵੇ:


```python
from scipy.stats import ttest_ind

tval, pval = ttest_ind(df.loc[df['Role']=='First_Baseman',['Height']], df.loc[df['Role']=='Second_Baseman',['Height']],equal_var=False)
print(f"T-value = {tval[0]:.2f}\nP-value: {pval[0]}")
```

`ttest_ind` ਫੰਕਸ਼ਨ ਵੱਲੋਂ ਵਾਪਸ ਕੀਤੀਆਂ ਦੋ ਮੁੱਲਾਂ ਹਨ:
* p-ਮੁੱਲ ਨੂੰ ਦੋ ਵੰਡਾਂ ਦੇ ਇਕੋ ਜਿਹੇ ਮਤਲਬ ਹੋਣ ਦੀ ਸੰਭਾਵਨਾ ਮੰਨੀ ਜਾ ਸਕਦੀ ਹੈ। ਸਾਡੇ ਕੇਸ ਵਿੱਚ, ਇਹ ਬਹੁਤ ਘੱਟ ਹੈ, ਜਿਸਦਾ ਅਰਥ ਹੈ ਕਿ ਇਹ ਮਜ਼ਬੂਤ ਸਬੂਤ ਹੈ ਕਿ ਪਹਿਲੇ ਬੇਸਮੈਨ ਊਚੇ ਹੁੰਦੇ ਹਨ।
* t-ਮੁੱਲ ਨਾਰਮਲਾਈਜ਼ਡ ਮਤਲਬ ਫਰਕ ਦੀ ਦਰਮਿਆਨੀ ਮੁੱਲ ਹੈ ਜੋ t-ਟੈਸਟ ਵਿੱਚ ਵਰਤੀ ਜਾਂਦੀ ਹੈ, ਅਤੇ ਇਸ ਦੀ ਤੁਲਨਾ ਦਿੱਤੇ ਗਏ ਭਰੋਸੇਮੰਦ ਮੁੱਲ ਲਈ ਕਰਮ ਤੋਂ ਕੀਤੀ ਜਾਂਦੀ ਹੈ।


## ਕੇਂਦਰੀ ਸੀਮਾ ਸਭੰਧੀ ਸਿਧਾਂਤ ਨਾਲ ਆਮ ਵੰਡ ਦਾ ਅਨੁਕਰਨ ਕਰਨਾ

ਪਾਈਥਨ ਵਿਚ ਛਦਮ-ਯਾਦਰਚ ਛੇਤੀ ਨਾਲ ਇੱਕ ਸਮਰੂਪ ਵੰਡ ਦੇਣ ਲਈ ਬਣਾਇਆ ਗਿਆ ਹੈ। ਜੇ ਅਸੀਂ ਸਧਾਰਣ ਵੰਡ ਲਈ ਇੱਕ ਜੇਨੇਰੇਟਰ ਬਣਾਉਣਾ ਚਾਹੁੰਦੇ ਹਾਂ, ਤਾਂ ਅਸੀਂ ਕੇਂਦਰੀ ਸੀਮਾ ਸਿਧਾਂਤ ਦਾ ਉਪਯੋਗ ਕਰ ਸਕਦੇ ਹਾਂ। ਇੱਕ ਸਧਾਰਣ ਵੰਡ ਵਾਲੀ ਮੂਲ ਨੂੰ ਪ੍ਰਾਪਤ ਕਰਨ ਲਈ ਅਸੀਂ ਬੱਸ ਸਮਰੂਪ-ਜਨਰੇਟ ਕੀਤੇ ਸੈੰਪਲ ਦਾ ਮੀਨ ਕੱਢਾਂਗੇ।


```python
def normal_random(sample_size=100):
    sample = [random.uniform(0,1) for _ in range(sample_size) ]
    return sum(sample)/sample_size

sample = [normal_random() for _ in range(100)]
plt.figure(figsize=(10,6))
plt.hist(sample)
plt.tight_layout()
plt.show()
```

## Correlation and Evil Baseball Corp

Correlation ਸਾਡੇ ਲਈ ਡਾਟਾ ਸੀਕੁਐਂਸز ਵਿਚਕਾਰ ਸੰਬੰਧ ਲੱਭਣ ਦੀ ਆਗਿਆ ਦਿੰਦੀ ਹੈ। ਸਾਡੇ ਖੇਡ ਉਦਾਹਰਨ ਵਿੱਚ, ਆਓ ਮੰਨ ਲਈਏ ਕਿ ਇੱਕ ਬੁਰਾ ਬੇਸਬਾਲ ਕਾਰਪੋਰੇਸ਼ਨ ਹੈ ਜੋ ਆਪਣੇ ਖਿਡਾਰੀਆਂ ਨੂੰ ਉਹਨਾਂ ਦੀ ਉਚਾਈ ਦੇ ਅਨੁਸਾਰ ਪੈਸਾ ਦਿੰਦਾ ਹੈ - ਜਿੰਨਾ ਵੱਡਾ ਖਿਡਾਰੀ ਹੁੰਦਾ ਹੈ, ਉੰਨਾ ਜ਼ਿਆਦਾ ਪੈਸਾ ਉਸਨੂੰ ਮਿਲਦਾ ਹੈ। ਮੰਨ ਲਈਏ ਕਿ ਬੇਸ ਸੈਲਰੀ $1000 ਹੈ, ਅਤੇ ਉਚਾਈ ਦੇ ਅਧਾਰ 'ਤੇ $0 ਤੋਂ $100 ਤੱਕ ਦਾ ਵਾਧੂ ਬੋਨਸ ਹੈ। ਅਸੀਂ MLB ਦੇ ਅਸਲੀ ਖਿਡਾਰੀਆਂ ਨੂੰ ਲਵਾਂਗੇ, ਅਤੇ ਉਹਨਾਂ ਦੀਆਂ ਕਲਪਨਾਤਮਕ ਸੈਲਰੀਆਂ ਦੀ ਗਣਨਾ ਕਰਾਂਗੇ:


```python
heights = df['Height'].fillna(method='pad')
salaries = 1000+(heights-heights.min())/(heights.max()-heights.mean())*100
print(list(zip(heights, salaries))[:10])
```

ਅਸੀਂ ਹੁਣ ਉਹਨਾਂ ਲੜੀਵਾਰਿਆਂ ਦੀ ਕੋਵੈਰੀਅੰਸ ਅਤੇ ਕੋਰਿਲੇਸ਼ਨ ਦੀ ਗਣਨਾ ਕਰਦੇ ਹਾਂ। `np.cov` ਸਾਨੂੰ ਇੱਕ ਕਿਹਾ ਜਾਂਦਾ ਹੈ **ਕੋਵੈਰੀਅੰਸ ਮੈਟ੍ਰਿਕਸ** ਦੇਵੇਗਾ, ਜੋ ਇੱਕ ਤੋਂ ਵੱਧ ਵੈਰੀਏਬਲਾਂ ਲਈ ਕੋਵੈਰੀਅੰਸ ਦਾ ਵਿਸਤਾਰ ਹੈ। ਕੋਵੈਰੀਅੰਸ ਮੈਟ੍ਰਿਕਸ $M$ ਦਾ ਤੱਤ $M_{ij}$ ਇਨਪੁਟ ਵੈਰੀਏਬਲ $X_i$ ਅਤੇ $X_j$ ਵਿੱਚ ਕੋਰਿਲੇਸ਼ਨ ਹੁੰਦਾ ਹੈ, ਅਤੇ ਡਾਇਗਨਲ ਮੁੱਲ $M_{ii}$ $X_{i}$ ਦੀ ਵੈਰੀਅੰਸ ਹੁੰਦੀ ਹੈ। ਇਸੇ ਤਰ੍ਹਾਂ, `np.corrcoef` ਸਾਨੂੰ **ਕੋਰਿਲੇਸ਼ਨ ਮੈਟ੍ਰਿਕਸ** ਦੇਵੇਗਾ।


```python
print(f"Covariance matrix:\n{np.cov(heights, salaries)}")
print(f"Covariance = {np.cov(heights, salaries)[0,1]}")
print(f"Correlation = {np.corrcoef(heights, salaries)[0,1]}")
```

ਇੱਕ ਸੰਬੰਧ ਜੋ 1 ਦੇ ਬਰਾਬਰ ਹੁੰਦਾ ਹੈ, ਦਾ ਅਰਥ ਹੈ ਕਿ ਦੋ ਬਦਲਿਆਂ ਵਿੱਚ ਇੱਕ ਮਜ਼ਬੂਤ **ਰੇਖੀ ਸੰਬੰਧ** ਹੈ। ਅਸੀਂ ਇੱਕ ਮੁੱਲ ਨੂੰ ਦੂਜੇ ਦੇ ਮੁਕਾਬਲੇ ਪਲਾਟ ਕਰਕੇ ਰੇਖੀ ਸੰਬੰਧ ਨੂੰ ਨਜ਼ਰਅੰਦਾਜ਼ ਕਰ ਸਕਦੇ ਹਾਂ:


```python
plt.figure(figsize=(10,6))
plt.scatter(heights,salaries)
plt.tight_layout()
plt.show()
```

ਆਓ ਵੇਖੀਏ ਕਿ ਕੀ ਹੁੰਦਾ ਹੈ ਜੇ ਸੰਬੰਧ ਰੇਖੀ ਨਹੀਂ ਹੈ। ਮੰਨ ਲਓ ਕਿ ਸਾਡੀ ਕੰਪਨੀ ਨੇ ਕਦਾਂ ਅਤੇ ਤਨਖ਼ਾਹਾਂ ਦਰਮਿਆਨ ਸਪਸ਼ਟ ਰੇਖੀ ਨਿਰਭਰਤਾ ਨੂੰ ਛੁਪਾਉਣ ਦਾ ਫੈਸਲਾ ਕੀਤਾ, ਅਤੇ ਫਾਰਮੂਲੇ ਵਿੱਚ ਕੁਝ ਗੈਰ-ਰੇਖੀਤਾ ਜਿਵੇਂ ਕਿ `sin` ਸ਼ਾਮਿਲ ਕੀਤਾ:


```python
salaries = 1000+np.sin((heights-heights.min())/(heights.max()-heights.mean()))*100
print(f"Correlation = {np.corrcoef(heights, salaries)[0,1]}")
```

ਇਸ ਮਾਮਲੇ ਵਿੱਚ, ਸਬੰਧ ਥੋੜ੍ਹਾ ਛੋਟਾ ਹੈ, ਪਰ ਇਹ ਫਿਰ ਵੀ ਕਾਫੀ ਉੱਚਾ ਹੈ। ਹੁਣ, ਸੰਬੰਧ ਨੂੰ ਹੋਰ ਘੱਟ ਸਪਸ਼ਟ ਬਣਾਉਣ ਲਈ, ਅਸੀਂ ਤਨਖਾਹ ਵਿੱਚ ਕੁਝ ਵਧੂ ਬੇਤਰਤੀਬੀ ਜੋੜ ਕੇ ਕੁਝ ਵਿਅਕਤੀਗਤ ਬਦਲਾਅ ਸ਼ਾਮਲ ਕਰਨਾ ਚਾਹੁੰਦੇ ਹਾਂ। ਆਓ ਦੇਖੀਏ ਕੀ ਹੁੰਦਾ ਹੈ:


```python
salaries = 1000+np.sin((heights-heights.min())/(heights.max()-heights.mean()))*100+np.random.random(size=len(heights))*20-10
print(f"Correlation = {np.corrcoef(heights, salaries)[0,1]}")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(heights, salaries)
plt.tight_layout()
plt.show()
```

> ਕੀ ਤੁਸੀਂ ਅਨੁਮਾਨ ਲਾ ਸਕਦੇ ਹੋ ਕਿ ਇਹ ਡੌਟਸ ਇਸ ਤਰ੍ਹਾਂ ਲੰਬਕਾਰੀ ਲਾਈਨਾਂ ਵਿੱਚ ਕਿਉਂ ਲੱਗਦੇ ਹਨ?

ਅਸੀਂ ਕ੍ਰਿਤ੍ਰਿਮ ਤਰੀਕੇ ਨਾਲ ਬਣਾਈ ਗਈ ਧਾਰਣਾ ਜਿਵੇਂ ਕਿ ਤਨਖਾਹ ਅਤੇ ਪਰਵੇਖਤ ਵੈਰੀਏਬਲ *ਉਚਾਈ* ਵਿੱਚ ਸਬੰਧ ਵੇਖਿਆ ਹੈ। ਆਓ ਦੇਖੀਏ ਕਿ ਦੋ ਪਰਵੇਖਤ ਵੈਰੀਏਬਲ, ਜਿਵੇਂ ਕਿ ਉਚਾਈ ਅਤੇ ਵਜ਼ਨ ਵੀ, ਇੱਕ ਦੂਜੇ ਨਾਲ ਸੰਬੰਧਿਤ ਹਨ ਜਾਂ ਨਹੀਂ:


```python
np.corrcoef(df['Height'].ffill(),df['Weight'])
```

ਬਦਕਿਸਮਤੀ ਨਾਲ, ਸਾਨੂੰ ਕੋਈ ਨਤੀਜਾ ਨਹੀਂ ਮਿਲਿਆ - ਸਿਰਫ ਕੁਝ ਅਜੀਬ `nan` ਮੁੱਲ। ਇਹ ਇਸ ਕਾਰਨ ਹੈ ਕਿ ਸਾਡੇ ਸੀਰੀਜ਼ ਵਿੱਚ ਕੁਝ ਮੁੱਲ ਅਣਨਿਰਧਾਰਿਤ ਹਨ, ਜੋ ਕਿ `nan` ਰੂਪ ਵਿੱਚ ਦਰਸਾਏ ਗਏ ਹਨ, ਜਿਸ ਕਾਰਨ ਓਪਰੇਸ਼ਨ ਦਾ ਨਤੀਜਾ ਵੀ ਅਣਨਿਰਧਾਰਿਤ ਹੋ ਜਾਂਦਾ ਹੈ। ਮੈਟਰਿਕਸ ਨੂੰ ਦੇਖ ਕੇ ਅਸੀਂ ਵੇਖ ਸਕਦੇ ਹਾਂ ਕਿ `Weight` ਸਮੱਸਿਆ ਵਾਲਾ ਕਾਲਮ ਹੈ, ਕਿਉਂਕਿ `Height` ਮੁੱਲਾਂ ਵਿੱਚ ਸਵੈ-ਸਬੰਧਤ ਕੈਲਕ्युਲੇਸ਼ਨ ਕੀਤਾ ਗਿਆ ਹੈ।

> ਇਹ ਉਦਾਹਰਨ **ਡੇਟਾ ਤਿਆਰੀ** ਅਤੇ **ਸਾਫ਼-ਸੁਥਰਾ** ਹੋਣ ਦੀ ਮਹੱਤਤਾ ਦਿਖਾਉਂਦੀ ਹੈ। ਠੀਕ ਡੇਟਾ ਤੋਂ ਬਿਨਾਂ ਅਸੀਂ ਕੁਝ ਵੀ ਗਣਨਾ ਨਹੀਂ ਕਰ ਸਕਦੇ।

ਚਲੋ `fillna` ਮੈਥਡ ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਗੁਆਚੁਕੇ ਮੁੱਲ ਭਰਦੇ ਹਾਂ, ਅਤੇ ਸਬੰਧਤਾ ਦੀ ਗਣਨਾ ਕਰਦੇ ਹਾਂ:


```python
np.corrcoef(df['Height'].fillna(method='pad'), df['Weight'])
```

ਇੱਕ ਸੰਬੰਧ ਹੈ, ਪਰ ਸਾਡੇ ਕਲਪਨਾਤਮਕ ਉਦਾਹਰਨ ਵਾਂਗੋਂ ਇਨਾ ਮਜ਼ਬੂਤ ਨਹੀਂ। ਵਿਅਕਤੀਗਤ ਮੀਮਾਂਸਾ ਦੇ ਲਈ ਜੇ ਅਸੀਂ ਇੱਕ ਮੁੱਲ ਨੂੰ ਦੂਜੇ ਮੁੱਲ ਦੇ ਖਿਲਾਫ ਸਕੈਟਰ ਪਲਾਟ 'ਚ ਵੇਖੀਏ, ਤਾਂ ਇਹ ਸੰਬੰਧ ਕਾਫੀ ਘੱਟ ਸਪਸ਼ਟ ਹੋਵੇਗਾ:


```python
plt.figure(figsize=(10,6))
plt.scatter(df['Weight'],df['Height'])
plt.xlabel('Weight')
plt.ylabel('Height')
plt.tight_layout()
plt.show()
```

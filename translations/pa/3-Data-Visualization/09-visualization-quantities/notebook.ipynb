# ਆਓ ਪੰਛੀਆਂ ਬਾਰੇ ਸਿੱਖੀਏ

## ਪੰਛੀਆਂ ਦੀ ਪਰਿਭਾਸ਼ਾ

ਪੰਛੀ ਉਹ ਜੀਵ ਹਨ ਜੋ ਅਕਸਰ ਉੱਡ ਸਕਦੇ ਹਨ ਅਤੇ ਜਿਨ੍ਹਾਂ ਦੇ ਪੰਖ ਹੁੰਦੇ ਹਨ। ਉਹਨਾਂ ਦੇ ਸਰੀਰ 'ਤੇ ਪੰਖ ਹੁੰਦੇ ਹਨ ਅਤੇ ਉਹ ਅੰਡੇ ਦੇ ਕੇ ਪ੍ਰਜਨਨ ਕਰਦੇ ਹਨ। 

[!NOTE] ਪੰਛੀਆਂ ਦੇ ਸਾਰੇ ਪ੍ਰਜਾਤੀਆਂ ਉੱਡ ਨਹੀਂ ਸਕਦੀਆਂ। ਉਦਾਹਰਣ ਵਜੋਂ, ਸ਼ਤਰਮੁਰਗ ਅਤੇ ਪੇਂਗੁਇਨ।

## ਪੰਛੀਆਂ ਦੀਆਂ ਮੁੱਖ ਵਿਸ਼ੇਸ਼ਤਾਵਾਂ

### ਪੰਖ
ਪੰਖ ਪੰਛੀਆਂ ਦੀ ਸਭ ਤੋਂ ਵੱਡੀ ਵਿਸ਼ੇਸ਼ਤਾ ਹੈ। ਇਹ ਉਹਨਾਂ ਨੂੰ ਉੱਡਣ ਵਿੱਚ ਮਦਦ ਕਰਦੇ ਹਨ ਅਤੇ ਸਰੀਰ ਨੂੰ ਗਰਮੀ ਤੋਂ ਬਚਾਉਂਦੇ ਹਨ।

### ਚੁੰਝ
ਪੰਛੀਆਂ ਦੀ ਚੁੰਝ ਉਹਨਾਂ ਨੂੰ ਖਾਣਾ ਖਾਣ ਵਿੱਚ ਮਦਦ ਕਰਦੀ ਹੈ। ਚੁੰਝ ਦਾ ਆਕਾਰ ਅਤੇ ਆਕਾਰ ਪੰਛੀ ਦੀ ਖੁਰਾਕ 'ਤੇ ਨਿਰਭਰ ਕਰਦਾ ਹੈ।

[!TIP] ਕੁਝ ਪੰਛੀਆਂ ਦੀ ਚੁੰਝ ਬਹੁਤ ਮਜ਼ਬੂਤ ਹੁੰਦੀ ਹੈ, ਜੋ ਉਹਨਾਂ ਨੂੰ ਸਖ਼ਤ ਬੀਜ ਤੋੜਨ ਵਿੱਚ ਮਦਦ ਕਰਦੀ ਹੈ।

### ਅੰਡੇ
ਪੰਛੀ ਅੰਡੇ ਦੇ ਕੇ ਆਪਣੇ ਬੱਚਿਆਂ ਨੂੰ ਜਨਮ ਦਿੰਦੇ ਹਨ। ਇਹ ਅੰਡੇ ਅਕਸਰ ਘੋਸਲੇ ਵਿੱਚ ਰੱਖੇ ਜਾਂਦੇ ਹਨ।

## ਪੰਛੀਆਂ ਦੇ ਵੱਖ-ਵੱਖ ਪ੍ਰਕਾਰ

### ਗਾਇਕ ਪੰਛੀ

### ਸ਼ਿਕਾਰੀ ਪੰਛੀ
ਇਹ ਪੰਛੀ ਮਾਸ ਖਾਂਦੇ ਹਨ ਅਤੇ ਸ਼ਿਕਾਰ ਕਰਨ ਵਿੱਚ ਮਾਹਰ ਹੁੰਦੇ ਹਨ। ਉਦਾਹਰਣ ਲਈ, ਬਾਜ ਅਤੇ ਗਿੱਧ।

[!WARNING] ਸ਼ਿਕਾਰੀ ਪੰਛੀਆਂ ਦੇ ਨਜ਼ਰ ਬਹੁਤ ਤੇਜ਼ ਹੁੰਦੀ ਹੈ। ਉਹ ਦੂਰ ਤੋਂ ਹੀ ਆਪਣੇ ਸ਼ਿਕਾਰ ਨੂੰ ਦੇਖ ਸਕਦੇ ਹਨ।

### ਪਾਣੀ ਵਾਲੇ ਪੰਛੀ

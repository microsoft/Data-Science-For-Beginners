## ನಂಬಿಕೆ ನಡುವಣಿಗಳು

ಈಗ ನಾವು ಬೇಸ್‌ಬಾಲ್ ಆಟಗಾರರ ತೂಕ ಮತ್ತು ಎತ್ತರಗಳಿಗಾಗಿ ನಂಬಿಕೆ ನಡುವೆಗಳನ್ನು ಲೆಕ್ಕ ಹಾಕೋಣ. ನಾವು ಈ ಕೋಡ್ [ಈ ಸ್ಟ್ಯಾಕ್ಓವರ್‌ಫ್ಲೋ ಚರ್ಚೆಯಿಂದ](https://stackoverflow.com/questions/15033511/compute-a-confidence-interval-from-sample-data) ಬಳಸಲಿದ್ದೇವೆ:


```python
import scipy.stats

def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, h

for p in [0.85, 0.9, 0.95]:
    m, h = mean_confidence_interval(df['Weight'].fillna(method='pad'),p)
    print(f"p={p:.2f}, mean = {m:.2f} ± {h:.2f}")
```

## ನಿಯಮಾನುಸ್ಥಾನ ಪರೀಕ್ಷೆ

ನಮ್ಮ ಬೇಸ್ಬಾಲ್ ಆಟಗಾರರ ಡೇಟಾಸೆಟ್‌ನಲ್ಲಿ ವಿಭಿನ್ನ ಪಾತ್ರಗಳನ್ನು ಪರಿಶೀಲಿಸೋಣ:


```python
df.groupby('Role').agg({ 'Weight' : 'mean', 'Height' : 'mean', 'Age' : 'count'}).rename(columns={ 'Age' : 'Count'})
```

ನಾವು ಮೊದಲ ಬೆಸ್ಮೆನ್‌ಗಳು ಎರಡನೇ ಬೆಸ್ಮೆನ್‌ಗಳಿಗಿಂತ ಎತ್ತರರಾಗಿದ್ದಾರೆ ಎಂಬ ಊಹೆಯನ್ನು ಪರೀಕ್ಷಿಸೋಣ. ಇದನ್ನು ಪರಿಶೀಲಿಸುವ ಅತ್ಯಂತ ಸರಳ ವಿಧಾನವು ನಂಬಿಕೆ ಅಂತರಗಳನ್ನು ಪರೀಕ್ಷಿಸುವುದಾಗಿದೆ:


```python
for p in [0.85,0.9,0.95]:
    m1, h1 = mean_confidence_interval(df.loc[df['Role']=='First_Baseman',['Height']],p)
    m2, h2 = mean_confidence_interval(df.loc[df['Role']=='Second_Baseman',['Height']],p)
    print(f'Conf={p:.2f}, 1st basemen height: {m1-h1[0]:.2f}..{m1+h1[0]:.2f}, 2nd basemen height: {m2-h2[0]:.2f}..{m2+h2[0]:.2f}')
```

ನಾವು интервалಗಳು ಮುರಿದಿಲ್ಲ ಎಂದು ನೋಡಬಹುದು.

ಸಂದರ್ಭಾತ್ಮಕವಾಗಿ hypothesesನ ಸರಿ ಎನ್ನುವ ವಿಶ್ಲೇಷಣೆಯ ಕೆಲವೊಂದು ಸರಿಯಾದ ವಿಧಾನವು **ಸ್ಟುಡೆಂಟ್ ಟಿ-ಟೆಸ್ಟ್** ಅನ್ನು ಬಳಸುವುದಾಗಿದೆ:


```python
from scipy.stats import ttest_ind

tval, pval = ttest_ind(df.loc[df['Role']=='First_Baseman',['Height']], df.loc[df['Role']=='Second_Baseman',['Height']],equal_var=False)
print(f"T-value = {tval[0]:.2f}\nP-value: {pval[0]}")
```

`ttest_ind` ಫಂಕ್ಷನ್ ನೀಡುವ ಎರಡು ಮೌಲ್ಯಗಳು ಇಂತಿವೆ:  
* p-ಮೌಲ್ಯವನ್ನು ಎರಡು توزيعات ಸಮಾನ ಸರಾಸರಿ ಹೊಂದಿರುವ ಸಾಧ್ಯತೆ ಎಂದು ಪರಿಗಣಿಸಬಹುದು. ನಮ್ಮ ಸಂದರ್ಭದಲ್ಲಿ, ಇದು ಬಹಳ ಕಡಿಮೆ, ಅಂದರೆ ಮೊದಲ baseman ಗಳು ಎತ್ತರವಾಗಿದ್ದಾರೆ ಎಂಬುದಕ್ಕೆ ಬಲವಾದ ಸಾಬೀತು ಇದೆ.  
* t-ಮೌಲ್ಯವು t-ಪರೀಕ್ಷೆಯಲ್ಲಿ ಬಳಸಲಾಗುವ ಸಾಮಾನ್ಯೀಕೃತ ಸರಾಸರಿ ವ್ಯತ್ಯಾಸದ ಮಧ್ಯಂತರ ಮೌಲ್ಯವಾಗಿದೆ ಮತ್ತು ನಿರ್ಧಿಷ್ಟ ನಂಬಿಕೆಯ ಮೌಲ್ಯಕ್ಕಾಗಿ ದಡ ಮೌಲ್ಯಕ್ಕೆ ಹೋಲಿಸಲಾಗುತ್ತದೆ.


## ಕೇಂದ್ರ ಸೀಮಿತ ಸಿದ್ಧಾಂತంతో ಸಾಮಾನ್ಯ ವಿತರಣೆಯನ್ನು ಅನೂಕಲಿಸು

ಪೈಥಾನ್‌ನ ಪ್ಸ್‌ಯುಡೋ-ಯಾದೃಚ್ಛಿಕ ಜನಕ ನಮಗೆ ಒರಟು ವಿತರವನ್ನು ನೀಡುವಂತೆ ವಿನ್ಯಾಸಗೊಳಿಸಲಾಗಿದೆ. ನಾವು ಸಾಮಾನ್ಯ ವಿತರಣೆಗಾಗಿ ಜನಕವನ್ನು ರಚಿಸಲು ಇಚ್ಛಿಸಿದರೆ, ನಾವು ಕೇಂದ್ರ ಸೀಮಿತ ಸಿದ್ಧಾಂತವನ್ನು ಬಳಸಬಹುದು. ಸಾಮಾನ್ಯವಾಗಿ ವಿತರಿಸಲಾದ ಮೌಲ್ಯವನ್ನು ಪಡೆಯಲು ನಾವು ಒರಟು-ಉತ್ಪನ್ನಿತ ಮಾದರಿಯ ಸರಾಸರಿಯನ್ನು ಲೆಕ್ಕಿಸುವುದೇ ಆಗಿದೆ.


```python
def normal_random(sample_size=100):
    sample = [random.uniform(0,1) for _ in range(sample_size) ]
    return sum(sample)/sample_size

sample = [normal_random() for _ in range(100)]
plt.figure(figsize=(10,6))
plt.hist(sample)
plt.tight_layout()
plt.show()
```

## ಪರಸ್ಪರ ಸಂಬಂಧ ಮತ್ತು ಇವರು ಬ್ಲೇಸ್‌ಬಾಲ್ ಕಾರ್ಪ್

ಪರಸ್ಪರ ಸಂಬಂಧವು ನಮಗೆ ಡೇಟಾ ಕ್ರಮಗಳ ನಡುವಿನ ಸಂಬಂಧಗಳನ್ನು ಕಂಡುಹಿಡಿಯಲು ಸಹಾಯ ಮಾಡುತ್ತದೆ. ನಮ್ಮ ಆಟದ ಉದಾಹರಣೆಯಲ್ಲಿ, ಕ್ರೀಡಾಪಟುಗಳ ಎತ್ತರದ ಪ್ರಕಾರ ಪಾವತಿ ನೀಡುವ ಒಂದು ಕೆಟ್ಟ ಬ್ಲೇಸ್‌ಬಾಲ್ ಸಂಸ್ಥೆ ಇದೆ ಎಂದು ಕಲ್ಪಿಸೋಣ - ಎತ್ತರದ ಆಟಗಾರನು ಆಗಿದ್ದರೆ, ಅವರು ಹೆಚ್ಚು ಹಣ ಪಡೆಯುತ್ತಾರೆ. ಆಧಾರ ವೇತನವು $1000 ಆಗಿದ್ದು, ಎತ್ತರದ ಪ್ರಕಾರ $0 ರಿಂದ $100 ರವರೆಗೂ ಹೆಚ್ಚುವರಿ ಬಹುಮಾನ ಇದೆ ಎಂದು ಕಲ್ಪಿಸೋಣ. ನಾವು MLB ನಿಂದ ನಿಜವಾದ ಆಟಗಾರರನ್ನು ತೆಗೆದುಕೊಳ್ಳುತ್ತೇವೆ ಮತ್ತು ಅವರ ಕಲ್ಪನಾತ್ಮಕ ವೇತನಗಳನ್ನು ಲೆಕ್ಕಿಸುತ್ತೇವೆ:


```python
heights = df['Height'].fillna(method='pad')
salaries = 1000+(heights-heights.min())/(heights.max()-heights.mean())*100
print(list(zip(heights, salaries))[:10])
```

ಈಗ ಆ ಸರಣಿಗಳ ಸಹಪ್ರಯೋಗ ಮತ್ತು ಸಹಸಂಬಂಧವನ್ನು ಗಣನೆ ಮಾಡೋಣ. `np.cov` ನಮಗೆ ಸಹಪ್ರಯೋಗ ಮ್ಯಾಟ್ರಿಕ್ಸ್ ಅನ್ನು ಒದಗಿಸುತ್ತದೆ, ಇದು ಬಹುಮಾರ್ಗಗಳೊಂದಿಗೆ ಸಹಪ್ರಯೋಗದ ವಿಸ್ತರಣೆ ಆಗಿದೆ. ಸಹಪ್ರಯೋಗ ಮ್ಯಾಟ್ರಿಕ್ಸ್ $M$ ರ $M_{ij}$ ಅಂಶವು ಇನ್‌ಪುಟ್ ವ್ಯತ್ಯಾಸಗಳು $X_i$ ಮತ್ತು $X_j$ ನಡುವೆ ಇರುವ ಸಂಬಂಧವಾಗಿದೆ, ಮತ್ತು ಡಯಾಗನಲ್ ಮೌಲ್ಯಗಳು $M_{ii}$ $X_i$ ರ ವೈವಿಧ್ಯತೆಯಾಗಿದೆ. ಅದೇ ರೀತಿ, `np.corrcoef` ನಮಗೆ **ಸಹಸಂಬಂಧ ಮ್ಯಾಟ್ರಿಕ್ಸ್** ಅನ್ನು ಒದಗಿಸುತ್ತದೆ.


```python
print(f"Covariance matrix:\n{np.cov(heights, salaries)}")
print(f"Covariance = {np.cov(heights, salaries)[0,1]}")
print(f"Correlation = {np.corrcoef(heights, salaries)[0,1]}")
```

ಒಂದು ಸಹಸಂಬಂಧ 1ಕ್ಕೆ ಸಮನಾಗಿದ್ದರೆ, ಅದು ಎರಡು ಚರಗಳ ನಡುವೆ ಬಲವಾದ **ರೇಖೀಯ ಸಂಬಂಧ** ಇದೆ ಎಂದು ಅರ್ಥ. ನಾವು ಒಂದು ಮೌಲ್ಯವನ್ನು ಇನ್ನೊಂದರ ವಿರುದ್ಧ ರೇಖೀಯ ಸಂಬಂಧವನ್ನು ದೃಶ್ಯರೀತಿಯಲ್ಲಿ ನೋಡಬಹುದು:


```python
plt.figure(figsize=(10,6))
plt.scatter(heights,salaries)
plt.tight_layout()
plt.show()
```

ಸಂಬಂಧ ರೇಖೀಯವಾಗದಿದ್ದರೆ ಏನು ಕಾಣಬಹುದು ಎಂದು ನೋಡೋಣ. ಎಣಿಸೋಣ ನಮ್ಮ ಸಂಸ್ಥೆ ಎತ್ತರ ಮತ್ತು ವೇತನಗಳ ನಡುವಿನ ಸ್ಪಷ್ಟ ರೇಖೀಯ ಅವಲಂಬನೆಯನ್ನು ಸ್ಪಷ್ಟವಾಗಿ ಮರೆಮಾಚಲು ನಿರ್ಧರಿಸಿದೆ ಮತ್ತು ಸೂತ್ರದಲ್ಲಿ ಕೆಲವು ಅರೆ-ರೇಖೀಯತೆಯನ್ನು, ಉದಾಹರಣೆಗೆ `sin` ಅನ್ನು ಪರಿಚಯಿಸಿದೆ:


```python
salaries = 1000+np.sin((heights-heights.min())/(heights.max()-heights.mean()))*100
print(f"Correlation = {np.corrcoef(heights, salaries)[0,1]}")
```

ಈ ಸಂದರ್ಭದಲ್ಲಿ, ಸಹಸಂಬಂಧ ಸ್ವಲ್ಪ ಕಡಿಮೆ ಆದರೆ ಅದು ಇನ್ನೂ ತುಂಬಾ ಎತ್ತರವಾಗಿದೆ. ಈಗ, ಸಂಬಂಧವನ್ನು ಇನ್ನಷ್ಟು ಅಸ್ಪಷ್ಟಗೊಳಿಸಲು, ವೇತನಕ್ಕೆ ಕೆಲವು ಹೆಚ್ಚಿನ ಯಾದೃಚ್ಛಿಕತೆಗಳನ್ನು ಸೇರಿಸುವ ಮೂಲಕ ನಾವು ಕೆಲವು ಯಾದೃಚ್ಛಿಕ ವೇರಿಯಬಲ್ ಸೇರಿಸಬಹುದು. ಏನು ಆಗುತ್ತದೆ ನೋಡೋಣ:


```python
salaries = 1000+np.sin((heights-heights.min())/(heights.max()-heights.mean()))*100+np.random.random(size=len(heights))*20-10
print(f"Correlation = {np.corrcoef(heights, salaries)[0,1]}")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(heights, salaries)
plt.tight_layout()
plt.show()
```

> ನೀವು ಯಾಕೆ ಈ ಬಿಂದುಗಳು ಇಂಡುಕಾರಿಯಾಗಿ ನಿಂತಿರುವ ಎಂದು ಊಹಿಸಬಹುದೆ?

ನಾವು ಸಂಬಳ ಮುಂತಾದ ಕೃತಕವಾಗಿ ನಿರ್ದಿಷ್ಟಪಡಿಸಿದ ಪರಿಕಲ್ಪನೆ ಮತ್ತು ಗಮನಿಸಲ್ಪಟ್ಟ ಚರ *ದೈರ್ಧ್ಯ* ನಡುವಿನ ಅನುಪಾತವನ್ನು ಗಮನಿಸಿದ್ದೇವೆ. ಹೀಗೆ, ಬಗೆಯ ಮತ್ತು ತೂಕ ಮುಂತಾದ ಎರಡೂ ಗಮನಿಸಲ್ಪಟ್ಟ ಚರಗಳು ಸಹ ಪರಸ್ಪರ ಸಂಬಂಧ ಹೊಂದಿದೆಯೇ ಎಂದು ನೋಡೋಣ:


```python
np.corrcoef(df['Height'].ffill(),df['Weight'])
```

ದುರುದೃಷ್ಟವಶಾತ್, ನಮಗೆ ಯಾವುದೇ ಫಲಿತಾಂಶ ಸಿಗಲಿಲ್ಲ - শুধুমಾತ್ರ ಸ್ವಲ್ಪ ವಿಚಿತ್ರ `nan` ಮೌಲ್ಯಗಳಿವೆ. ನಮ್ಮ ಸರಣಿಯಲ್ಲಿ ಕೆಲವು ಮೌಲ್ಯಗಳು ನಿರ್ಧರಿಸದಿರುವುದರಿಂದ, ಅವು `nan` ಆಗಿ ಪ್ರತಿನಿಧಿಸಲಾಗಿದ್ದು, ಇದರಿಂದ ಕಾರ್ಯಾಚರಣೆಯ ಫಲಿತಾಂಶವೂ ನಿರ್ಧರಿಸದಿರುತ್ತದೆ. ಮ್ಯಾಟ್ರಿಕ್ಸ್ ನೋಡಿದಾಗ ನಾವು ಕಾಣಬಹುದು `Weight` ಸಮಸ್ಯೆಯಿರುವ ಕಾಲಮ್, ಏಕೆಂದರೆ `Height` ಮೌಲ್ಯಗಳ ನಡುವಿನ ಸ್ವ-ಸಂಬಂಢನ ಲೆಕ್ಕಹಾಕಲಾಗಿದೆ.

> ಈ ಉದಾಹರಣೆ **ಡೇಟಾ ತಯಾರಿ** ಮತ್ತು **ಶುದ್ಧೀಕರಣ** ಮಹತ್ವವನ್ನು ತೋರಿಸುತ್ತದೆ. ಸರಿಯಾದ ಡೇಟಾ ಇಲ್ಲದೆ ನಾವು ಯಾವುದನ್ನೂ ಲೆಕ್ಕಹಾಕಲು ಸಾಧ್ಯವಿಲ್ಲ.

ನಿಮ್ಮ ಕಳೆದುಹೋಗಿರುವ ಮೌಲ್ಯಗಳನ್ನು ತುಂಬಲು `fillna` ವಿಧಾನವನ್ನು ಬಳಸಿ, ಮತ್ತು ಸಂಬಂಧ ಲೆಕ್ಕಹಾಕೋಣ:


```python
np.corrcoef(df['Height'].fillna(method='pad'), df['Weight'])
```

ನಿಜಕ್ಕೂ ಒಂದು ಸಂಬಂಧವಿದೆ, ಆದರೆ ನಮ್ಮ ಕೃತಕ ಉದಾಹರಣೆಯಂತೆ ಬಲವಾದುದಾಗಿಲ್ಲ. ನಿಜಕ್ಕೂ, ಒಂದು ಮೌಲ್ಯವನ್ನು ಇನ್ನೊಂದರ ವಿರುದ್ಧದ ಚಿತ್ರ ಸಮೀಕ್ಷೆಯನ್ನು ನೋಡಿದರೆ, ಸಂಬಂಧವು ಸ್ಪಷ್ಟವಾಗಿರುವುದಿಲ್ಲ:


```python
plt.figure(figsize=(10,6))
plt.scatter(df['Weight'],df['Height'])
plt.xlabel('Weight')
plt.ylabel('Height')
plt.tight_layout()
plt.show()
```
